```python
df=df.drop_duplicates() #loại bỏ các giá trị trùng lặp
```

```python
#Làm sạch dữ liệu
df.isna()
```

```python
df[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI","Age"]] = df[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI","Age"]].replace(0, np.NaN)
#Pregnancies có thể =0 nên không cần thay
```

```python
df.isna().sum()
#đếm dữ liệu NaN
```

```python
df["Glucose"].fillna(df["Glucose"].mean(), inplace = True)
df["BloodPressure"].fillna(df["BloodPressure"].mean(), inplace = True)
df["SkinThickness"].fillna(df["SkinThickness"].mean(), inplace = True)
df["Insulin"].fillna(df["Insulin"].mean(), inplace = True)
df["BMI"].fillna(df["BMI"].mean(), inplace = True)
#thay thế giá trị NaN bằng số trung vị của cột đó
```

```python
df.hist(bins=10,figsize=(10,10))
plt.show()
```

NX:chỉ có Glucose và BloodPressure tuân theo phân phối chuẩn

```python
df.head(10)
```

```python
#Xac dinh thuoc tinh mo ta X va thuoc tinh du doan y
features = ['Pregnancies', 'Glucose','BloodPressure', 'SkinThickness', 'Insulin','BMI',
            'DiabetesPedigreeFunction', 'Age']
target=['Outcome']
X= df[features]
y = df[target]
```

```python
#Chia bo du lieu thanh 2 tap train va test theo ty le 80% train, 20% test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
df_train = [X_train, y_train]
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

**Phân lớp rừng**

```python
#Khai báo mô hình phân lớp rừng
from sklearn.ensemble import RandomForestClassifier 
#Xây dựng mô hình
model = RandomForestClassifier(n_estimators = 100, 
                               criterion = "gini", 
                               max_depth =3,
                               random_state = 1)
```

```python
#Thực thi mô hình trên tập train
model.fit(X_train, y_train) 
```

```python
#Kiem thu mo hinh tren tap test
y_pred = model.predict(X_test) 
print("Giá trị nhãn mô hình dự đoán được:\n",y_pred)
```

**HIỂN THỊ MÔ HÌNH**

```python
#Kết quả 5 cây quyết định từ mô hình phân lớp rừng
import graphviz
from graphviz import Source
from sklearn.tree import export_graphviz
for i in range(5):
    tree = model.estimators_[i]
    dot_data = export_graphviz(tree,
                               feature_names=features,  
                               class_names = ["Disease", "No Disease"],
                               filled=True,  
                               max_depth=3,
                               impurity=True, 
                               proportion=True)
    graph = graphviz.Source(dot_data)
    display(graph)
    graph.format = 'png'
    graph.render(f'random_forest_tree_{i}', view=False)
```

```python
#Xác định ma trận nhầm lẫn
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred))
disp = ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred))
disp.plot()
```

```python
#Độ chính xác của mô hình
print("Accuracy:", accuracy_score(y_test, y_pred)*100)
```

```python
#Ma trận độ chính xác của mô hình
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Glucose 120, Pregnancies 5, BloodPressure 75, SkinThickness 38, Insulin 100, BMI 29, DiabetesPedigreeFunction 0.2, Age 40
x=[[5, 180, 75, 38, 100, 29, 0.8, 40]]
y = model.predict(x)
print (y)
```

```python
#Vẽ mô hình barchart
import seaborn as sns
import matplotlib.pyplot as plt
importances = model.feature_importances_
features_importances = pd.Series(importances, index = X.columns).sort_values(ascending = False)
sns.barplot (x = features_importances, y = features_importances.index)
plt.xlabel ('Feature importances')
plt.ylabel ('Feature')
plt.show()
```

Có thể thấy rằng Glucose, BMI có mức độ tương quan cao nhất với kết quả. BloodPressure và DiabetesPedigreeFunction có mức độ tương quan thấp nhất.
